## Algo Palindrome 

```python
from unidecode import unidecode
def est_palindrome(mot):
    mot = mot.lower()  # Convertir en minuscules
    mot = ''.join(e for e in mot if e.isalnum())  # Supprimer espaces et ponctuation
    mot = unidecode(mot)  # Normaliser les caractères accentués
    
    longueur = len(mot)
    moitie = longueur // 2

    for i in range(moitie):
        if mot[i] != mot[longueur - i - 1]:
            return False

    return True
```

```python
# Exemple d'utilisation
mot_test = "A man, à plan, a canal, Panama!"
resultat = est_palindrome(mot_test)
print(resultat)  #
```

##anagramme

```python
def est_anagramme(mot1, mot2):
    mot1 = mot1.lower()
    mot2 = mot2.lower()

    # Supprimer espaces et ponctuation
    mot1 = ''.join(e for e in mot1 if e.isalnum())
    mot2 = ''.join(e for e in mot2 if e.isalnum())

    # Comparaison des fréquences de caractères
    def compter_frequence(mot):
        frequence = {}
        for char in mot:
            if char in frequence:
                frequence[char] += 1
            else:
                frequence[char] = 1
        return frequence

    return compter_frequence(mot1) == compter_frequence(mot2)
```

```python
# Exemple d'utilisation
mot1 = "listen"
mot2 = "silent"
resultat = est_anagramme(mot1, mot2)
print(resultat)  # Output: True
```
